```python
import numpy as np 

import tensorflow as tf
import keras

from keras.models import Sequential
from keras.layers import Input, Dense
```

```python
xs = np.array([-1, 0, 1, 2, 3, 4], dtype=np.float64)
ys = np.array([-3, 1, 1, 3, 5, 7], dtype=np.float64)
```

```python
model = Sequential([
    Input(shape=[1]),
    Dense(units=1)              # units = single artifical neuron
])
```

```python
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
model.fit(xs, ys, epochs=500)         #model training 
```

```python
model.predict(np.array([10]))            #single prediction
```
